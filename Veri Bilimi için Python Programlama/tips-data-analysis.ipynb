Seaborn ve pandas kütüphanelerini import ediyoruz.

```python
import seaborn as sns
import pandas as pd
```

Seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız.

```python
df2 = sns.load_dataset("tips")
df2.head()
```

Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill  değerlerinin toplamını, min, max ve ortalamasını bulunuz.

```python
df2.groupby("time").agg({"total_bill": ["sum", "min", "max", "mean"]})
```

Günlere ve time göre total_bill değerlerinin toplamını, min, max ve ortalamasını bulunuz.

```python
df2.groupby(["time", "day"]).agg({"total_bill": ["sum", "min", "max", "mean"]})
```

Lunch zamanına ve kadın müşterilere ait total_bill ve tip  değerlerinin day'e göre toplamını, min, max ve ortalamasını bulunuz.

```python
df2[(df2["time"] == "Lunch") & (df2["sex"] == "Female")].groupby("day").agg({"total_bill": ["sum", "min", "max", "mean"],
                                                                             "tip": ["sum", "min", "max", "mean"]})
```

size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir?

```python
df2.loc[(df2["size"] < 3) & (df2["total_bill"] > 10), "total_bill"].mean()
```

total_bill_tip_sum adında yeni bir değişken oluşturun. Her bir müşterinin ödediği totalbill ve tip in toplamını versin.

```python
df2["total_bill_tip_sum"] = df2["total_bill"] + df2["tip"]
df2.head()
```

total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız.

```python
new_df = df2.sort_values("total_bill_tip_sum", ascending=False)[:30]
new_df
```
