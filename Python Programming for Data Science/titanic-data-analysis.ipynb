deck değikenindeki boş değerleri deck değişkenin en çok tekrar eden değeri (mode) ile doldurunuz.

```python
# Pandas Serisi olduğu için, index belirtmemiz gerekiyor.
df["deck"].mode()[0] # Sonuç: 0    C
df["deck"].fillna(df["deck"].mode()[0])
df["deck"].isnull().sum()
```

age değikenindeki boş değerleri age değişkenin medyanı ile doldurun.

```python
df["age"].fillna(df["age"].median())
df.isnull().sum()
```

survived değişkeninin Pclass ve Cinsiyet değişkenleri kırılımınında sum, count, mean değerlerini bulunuz.

```python
df.groupby(["pclass", "sex"]).agg({"survived": ["sum", "count", "mean"]})
```

30 yaşın altında olanlara 1, 30'a eşit ve üstünde olanlara 0 verecek bir fonksiyon yazınız.
Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz oluşturunuz. (apply ve lambda yapılarını kullanınız)

```python
def age_group(age):
    if age < 30:
        return 1
    else:
        return 0

df["age_flag"] = df["age"].apply(lambda x: age_group(x))

df.head()

# 2. çözüm yolu

df["age_flag"] = df["age"].apply(lambda x: 1 if x < 30 else 0)
```
